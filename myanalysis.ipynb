```python
# Turn off the max column width so the images won't be truncated
pd.set_option('display.max_colwidth', None)
 
# Turning off the max column will display all the data
# if gathering into sets / array we might want to restrict to a few items
pd.set_option('display.max_seq_items', 3)

# Define a function for creating sparklines
def sparkline(data, figsize=(4, 0.25), **kwargs):
    """
    creates a sparkline
    """
    from matplotlib import pyplot as plt
    import base64
    from io import BytesIO

    data = list(data)

    *_, ax = plt.subplots(1, 1, figsize=figsize, **kwargs)
    ax.plot(data)
    ax.fill_between(range(len(data)), data, len(data)*[min(data)], alpha=0.1)
    ax.set_axis_off()

    img = BytesIO()
    plt.savefig(img)
    plt.close()
    return '<img src="data:image/png;base64, {}" />'.format(base64.b64encode(img.getvalue()).decode())
```

```python
# Import pycountry, a library for obtaining ISO country codes (alpha_3)
import pycountry

# Define a function for searching for ISO country codes (alpha_3) based on country names 
def do_fuzzy_search(country):
    try:
        # Use search_fuzzy() method to search for the ISO code for the country
        result = pycountry.countries.search_fuzzy(country)
    except Exception:
        # If the search finds no result, return a missing value
        return np.nan
    else:
        # If the search finds a ISO code, return alpha_3
        return result[0].alpha_3
    
# Use list comprehension to create a map from country name (country) to its ISO code (do_fuzzy_search(country))
iso_map = {country: do_fuzzy_search(country) for country in country_df["Country"].unique()}

# Create a column named Country_Code for the ISO code based on the country name in the Country column
country_df["Country_Code"] = country_df["Country"].map(iso_map)

# Remove rows without an ISO country code
country_df = country_df.loc[~(country_df['Country_Code'].isna()),]
country_df
```

```python
import math
from scipy.stats import t

sample_df = toyota_df.sample(frac=0.1, replace=False, random_state=1)

# A confidence level of 95%
alpha = 1 - 0.95
sample_mean = sample_df['price'].mean()
sample_std = sample_df['price'].std()
n = sample_df.shape[0]
# The degree of freedom for a sample size of sample_df.shape[0]
dof = n - 1

t_one_tail = t.ppf(alpha/2, dof)
error_margin = t_one_tail * (sample_std / math.sqrt(n))
print('Sample Mean: '+str(sample_mean))
print('Interval Estimate of a Population Mean: ['+str(sample_mean-abs(error_margin))+', '+str(sample_mean+abs(error_margin))+']')
```

```python
# Calculate t_score by applying the t-statistic formula
t_score = (sample_mean - 34790) / (sample_std / math.sqrt(n))
# Upper-tailed t-test for the mean menu price of Michelin restaurants
p_value = t.sf(abs(t_score), n-1)
p_value
```

```python
def forecast_accuracy(df, obs, pred):
    '''
    This function calculates four types of forecast errors.
    
    Parameters
    ----------
    df : DataFrame
        Container of observations and predictions.
    obs : str
        Column index for the observations.
    pred : str
        Column index for the predictions.

    Returns
    -------
    str
        A summary of all four types of forecast errors.
    '''
    mfe = (df[obs] - df[pred]).mean()
    mae = abs(df[obs] - df[pred]).mean()
    mse = ((df[obs] - df[pred]) ** 2).mean()
    mape = (((df[obs] - df[pred]) / df[obs]) * 100).mean()

    return 'Mean Forcasting Error (MFE): '+str(mfe)+\
           '\nMean Absolute Error (MAE): '+str(mae)+\
           '\nMean Squared Error (MSE): '+str(mse)+\
           '\nMean Absolute Percentage Error (MAPE): '+str(mape)
```
